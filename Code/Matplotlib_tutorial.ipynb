```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('test_data/stinkbug') / 255
print(img)
```

```python
plt.imshow(img)
plt.colorbar()
```

```python
gray = img[:, :, 1]
plt.imshow(gray)
plt.colorbar()
```

```python
plt.hist(gray.flatten(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```

```python
imgplot = plt.imshow(gray)
imgplot.set_clim(0.0, 0.7)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(gray)
ax.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(gray)
imgplot.set_clim(0.0, 0.7)
ax.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```
