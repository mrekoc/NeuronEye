```python
import matplotlib.pyplot as plt   #We use matplotlib library for plotting
import matplotlib.image as mpimg  
import matplotlib.colors as clr
import numpy as np                
```

```python
img = mpimg.imread("test_data/screws_375.png") 
img_hsv = clr.rgb_to_hsv(img)
#print(img_hsv[:,:,1])  
```

```python
img_plot= plt.imshow(img)
plt.imshow(img_hsv[:,:,1])
plt.colorbar()
```

```python
green_img = img [:, :, 1] * 255
img_sat = img_hsv[:,:,1] * 255
plt.imshow(green_img)
#print(green_img)
```

```python
hist,bins = np.histogram(img_sat.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(img_sat.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
imgplot = plt.imshow(img_sat)
imgplot.set_clim(0, 35)  
'''using saturation from the HSV colour model we can filter brightest objects
which are metal nuts in this case'''
```
