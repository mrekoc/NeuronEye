```python
import matplotlib.pyplot as plt   #We use matplotlib library for plotting
import matplotlib.image as mpimg  
import matplotlib.colors as clr
import numpy as np   
import cv2 as cv
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
img = cv.imread("test_data/screws_375.png",0) 
img = cv.resize(img, (480,360))
```

```python
img_plot= plt.imshow(img, cmap = 'gray')
plt.colorbar()
```

```python
plt.imshow(img, cmap = 'gray')
plt.colorbar()
```

```python
hist,bins = np.histogram(img.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
imgplot = plt.imshow(img)
#imgplot.set_clim(200, 250)  
'''using blue channel of an image and its corresponding histogram graph we can get meaningful results..'''
```

```python
_, thresh = cv.threshold(img, 250, 255, cv.THRESH_BINARY)

fig, axs = plt.subplots(1,2)
#fig.suptitle('Median bluring')
axs[0].imshow(img, cmap = 'gray')
axs[0].set_title('Original image')
axs[1].imshow(thresh, cmap = 'gray')
axs[1].set_title('Thresholded image')
```
